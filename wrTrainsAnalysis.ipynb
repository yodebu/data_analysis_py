Let us do some analysis with From_Time and To_Time.

```python
import dateutil,datetime
tData = data
tData['From_Time'] = tData['From_Time'].apply(dateutil.parser.parse)
tData['To_Time'] = tData['To_Time'].apply(dateutil.parser.parse)
tData.head()
hour = []
ft = tData['From_Time']
for rec in ft:
    hour.append(rec.hour)
tData['From_Hour'] = hour      
                             
```

# No of trains (irrespective of origin station) running by hour of the day.

```python
tData.groupby('From_Hour')['Train_No'].count().plot(kind='bar')
```

### From the above graph it is clear that 7 am to 10 am and 5 pm to 9 pm forms the peak period for commutation with minimum 1 train originating every minute. There is no train service between 1 am to 3 am.
